```python
import matplotlib.pyplot as plt
plt.title('legend') #제목 설정
#그래프 그리기
plt.plot([10,20,30,40],color='skyblue' ,label = 'asc') #증가를 의미하는 asc범례
plt.plot([40,30,20,10],'pink',label = 'desc') #감소를 의미하는 desc범례
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle') #제목 설정
#빨간색 dashed 그래프
plt.plot([10,20,30,40],color='r',linestyle='--',label='dashed')
#초록색 dotted 그래프
plt.plot([40,30,20,10],color='g',ls=':',label='dotted')
plt.legend() #범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker') #제목 설정
plt.plot([10,20,30,40],'r.',label='circle') #빨간색 원형 마커 그래프
#초록색 삼각형 마커 그래프
plt.plot([40,30,20,10],'g^',label='triangle up')
plt.legend() #범례표시
plt.show()
```
